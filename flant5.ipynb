```python
from transformers import T5Tokenizer, T5ForConditionalGeneration

tokenizer = T5Tokenizer.from_pretrained("google/flan-t5-xl")
model = T5ForConditionalGeneration.from_pretrained("google/flan-t5-xl", device_map="auto")
```

```python
PROMPT = '''Here are question answer pairs 

Q: when did the first harry potter book come out?
A: in 1997

Q: where were the beach scenes in chitty chitty bang bang filmed
A: Cap Taillat in St. Tropez , France
'''

def get_answer(question):
  input_text = PROMPT + f"\nQ: {question}?\nA: "
  input_ids = tokenizer(input_text, return_tensors="pt").input_ids.to("cuda")

  outputs = model.generate(input_ids)
  raw_answer = tokenizer.decode(outputs[0])

  answer = raw_answer.replace("<pad>", "").replace("</s>", "").strip()
  return answer
```

```python
import pandas as pd

data_pd = pd.read_csv("base_answer_pairs.csv")

for index, row in data_pd.iterrows():
    if index > 100:
        break
    data_pd.at[index, 'FLAN-T5 Answer'] = get_answer(row['Questions'])
    data_pd.to_csv("base_answer_pairs.csv", index=False)

data_pd.to_csv("base_answer_pairs.csv", index=False)
```
